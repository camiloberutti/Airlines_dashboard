## Dataset content:

- FL_DATE:
The date of the flight.Instead of storing all attributes to store date we use a single attribute.Date of the Flight in yyyymmdd Airline Reporting Unique.
- AIRLINE_ID:
An identification number assigned by US DOT to identify a unique airline (carrier). It has a single value for each individual carrier.
- TAIL_NUM:
An identification number to store the tail number of the airlines.
- FLIGHT_NUM:
The flight number is stored in form of unique value for each flight.
- ORIGIN_SEQ_ID:
Unique id for storing each origin airport of the flight.
- ORIGIN_AIRPORT:
Stores the origin airport.It can be used for displaying the airport which are easier to understand. For example, JFK.
- DEST_SEQ_ID:
Unique id for storing each destination airport of the flight.
- DEST_AIRPORT:
Stores the destination airport.It can be used for displaying the airport which are easier to understand.For example, JFK.
- DEP_TIME:
Actual Departure Time in local time: hhmm

Airline suggester:

```python
def suggest_airlines(df, origin_airport, dest_airport):
    """
    Suggests airlines for a given origin and destination airport based on average delay and flight frequency.

    Args:
        df (pd.DataFrame): The main flight data DataFrame.
        origin_airport (str): The IATA code of the origin airport.
        dest_airport (str): The IATA code of the destination airport.

    Returns:
        pd.DataFrame: A DataFrame with airline suggestions, sorted by average delay.
    """
    # Filter data for the specified route
    route_df = df[(df['ORIGIN_AIRPORT'] == origin_airport) & (df['DEST_AIRPORT'] == dest_airport)].copy()

    if route_df.empty:
        print(f"No flights found for the route: {origin_airport} to {dest_airport}")
        return None

    # Calculate average delay per airline
    airline_delays = route_df.groupby('AIRLINE_ID')['ARR_DELAY'].mean().reset_index(name='average_delay')

    # Calculate daily and weekly trip counts
    route_df['FL_DATE'] = pd.to_datetime(route_df['FL_DATE'])
    route_df['week'] = route_df['FL_DATE'].dt.isocalendar().week
    route_df['day_of_week'] = route_df['FL_DATE'].dt.dayofweek # Monday=0, Sunday=6


    weekly_trips = route_df.groupby(['AIRLINE_ID', 'week']).size().reset_index(name='weekly_count')
    avg_weekly_trips = weekly_trips.groupby('AIRLINE_ID')['weekly_count'].mean().reset_index(name='average_weekly_trips')

    daily_trips = route_df.groupby(['AIRLINE_ID', 'FL_DATE']).size().reset_index(name='daily_count')
    avg_daily_trips = daily_trips.groupby('AIRLINE_ID')['daily_count'].mean().reset_index(name='average_daily_trips')


    # Merge delay and trip counts
    airline_suggestions = pd.merge(airline_delays, avg_weekly_trips, on='AIRLINE_ID')
    airline_suggestions = pd.merge(airline_suggestions, avg_daily_trips, on='AIRLINE_ID')


    # Sort by average delay
    airline_suggestions = airline_suggestions.sort_values(by='average_delay')

    # Optional: Add airline name if available (requires merging with airline data)
    # For now, we use AIRLINE_ID. You would need a separate dataset for airline names.

    return airline_suggestions

# Example Usage:
# suggest_airlines(df, 'JFK', 'LAX')
```

```python
import plotly.express as px

def visualize_flight_counts(flight_counts_df, time_granularity='month', plot_type='bar'):
    """
    Visualizes flight counts per time period for airlines using a bar plot, line plot, or area plot.

    Args:
        flight_counts_df (pd.DataFrame): DataFrame with flight counts per time period
                                        (output of analyze_flights_by_time).
        time_granularity (str): The time period used for grouping ('year', 'month', or 'day').
        plot_type (str): The type of plot to generate ('bar', 'line', or 'area').
    """
    if flight_counts_df is None or flight_counts_df.empty:
        print("No data to visualize.")
        return

    # Convert time_period to datetime for better x-axis handling in plots
    flight_counts_df['time_period_dt'] = flight_counts_df['time_period'].dt.to_timestamp()


    if plot_type == 'bar':
        fig = px.bar(flight_counts_df,
                     x='time_period_dt',
                     y='flight_count',
                     color='AIRLINE_ID',
                     title=f'Flight Counts per {time_granularity.capitalize()} by Airline (Bar Chart)')
    elif plot_type == 'line':
         fig = px.line(flight_counts_df,
                     x='time_period_dt',
                     y='flight_count',
                     color='AIRLINE_ID',
                     title=f'Flight Counts per {time_granularity.capitalize()} by Airline (Line Chart)')
    elif plot_type == 'area':
         fig = px.area(flight_counts_df,
                      x='time_period_dt',
                      y='flight_count',
                      color='AIRLINE_ID',
                      title=f'Total Flight Volume per {time_granularity.capitalize()} by Airline (Area Chart)')
    else:
        print("Invalid plot_type. Choose from 'bar', 'line', or 'area'.")
        return


    fig.update_layout(xaxis_title=f'{time_granularity.capitalize()} and Year',
                      yaxis_title='Number of Flights')

    fig.show()

# Example Usage:
# Visualize monthly flight counts for all airlines as a bar chart
# visualize_flight_counts(all_airlines_monthly_flights, time_granularity='month', plot_type='bar')

# Visualize yearly flight counts for specific airlines as a line chart
# visualize_flight_counts(specific_airlines_yearly_flights, time_granularity='year', plot_type='line')

# Visualize daily flight counts for a specific airline as a bar chart
# visualize_flight_counts(specific_airline_daily_flights, time_granularity='day', plot_type='bar')

# Example to show the line chart for top airlines (using data generated in cell 3f64b5c2)
# visualize_flight_counts(top_airlines_monthly_flights, time_granularity='month', plot_type='line')

# Example to show the area chart for all airlines monthly flight counts
# visualize_flight_counts(all_airlines_monthly_flights, time_granularity='month', plot_type='area')
```

```python
# We have already identified the top 5 airlines by ID in a previous step (top_airlines_ids)
# If not, you can recalculate:
# n_top_airlines = 5
# airline_flight_counts = df['AIRLINE_ID'].value_counts().reset_index(name='total_flights')
# airline_flight_counts.rename(columns={'index': 'AIRLINE_ID'}, inplace=True)
# top_airlines_ids = airline_flight_counts['AIRLINE_ID'].head(n_top_airlines).tolist()
# print(f"Top {n_top_airlines} airlines IDs: {top_airlines_ids}")

# Ensure FL_DATE is datetime for time analysis
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])

# Create a new column to categorize airlines as 'Top N' or 'Other'
df['Airline_Category'] = df['AIRLINE_ID'].apply(lambda x: str(x) if str(x) in top_airlines_ids else 'Other')

def analyze_flights_by_time_categorized(df, time_granularity='month'):
    """
    Analyzes the number of flights per time period for 'Top N' airlines and 'Other' airlines.

    Args:
        df (pd.DataFrame): The main flight data DataFrame with 'Airline_Category'.
        time_granularity (str): The time period to group by ('year', 'month', or 'day').

    Returns:
        pd.DataFrame: A DataFrame with flight counts per time period for the categories.
    """
    df_filtered = df.copy()

    if df_filtered.empty:
        print("No data found for the specified criteria.")
        return None

    # Extract time components based on granularity
    if time_granularity == 'year':
        df_filtered['time_period'] = df_filtered['FL_DATE'].dt.to_period('Y')
    elif time_granularity == 'month':
        df_filtered['time_period'] = df_filtered['FL_DATE'].dt.to_period('M')
    elif time_granularity == 'day':
        df_filtered['time_period'] = df_filtered['FL_DATE'].dt.to_period('D')
    else:
        print("Invalid time_granularity. Choose from 'year', 'month', or 'day'.")
        return None

    # Group by airline category and time period and count flights
    flight_counts = df_filtered.groupby(['Airline_Category', 'time_period']).size().reset_index(name='flight_count')

    # Ensure 'Other' category is always present for consistency in stacking, even if count is 0 for a period
    all_time_periods = flight_counts['time_period'].unique()
    all_categories = flight_counts['Airline_Category'].unique()
    if 'Other' not in all_categories:
        all_categories = list(all_categories) + ['Other']

    # Create a complete index with all time periods and categories
    from itertools import product
    complete_index = pd.MultiIndex.from_product([all_categories, all_time_periods], names=['Airline_Category', 'time_period'])

    # Reindex and fill missing values with 0
    flight_counts = flight_counts.set_index(['Airline_Category', 'time_period']).reindex(complete_index, fill_value=0).reset_index()

    # Convert 'time_period' from Period to datetime for plotting
    flight_counts['time_period'] = flight_counts['time_period'].dt.to_timestamp()


    return flight_counts

# Analyze monthly flight counts for top 5 and 'Other' airlines
categorized_monthly_flights = analyze_flights_by_time_categorized(df, time_granularity='month')

# Visualize the monthly flight counts as a stacked bar chart with clearer colors
fig_monthly_categorized = px.bar(categorized_monthly_flights,
                                 x='time_period',
                                 y='flight_count',
                                 color='Airline_Category', # Use the new category for color
                                 title=f'Total Flight Volume per Month by Top {n_top_airlines} Airlines and Others',
                                 labels={'time_period': 'Month and Year', 'flight_count': 'Number of Flights'},
                                 color_discrete_sequence=px.colors.qualitative.Set1 # Use a clear categorical colorscale
                                 )
fig_monthly_categorized.update_layout(xaxis={'categoryorder':'category ascending'}) # Ensure chronological order
fig_monthly_categorized.show()


# Analyze daily flight counts for top 5 and 'Other' airlines
categorized_daily_flights = analyze_flights_by_time_categorized(df, time_granularity='day')

# Visualize the daily flight counts as a stacked bar chart with clearer colors
fig_daily_categorized = px.bar(categorized_daily_flights,
                                 x='time_period',
                                 y='flight_count',
                                 color='Airline_Category', # Use the new category for color
                                 title=f'Total Flight Volume per Day by Top {n_top_airlines} Airlines and Others',
                                 labels={'time_period': 'Date', 'flight_count': 'Number of Flights'},
                                 color_discrete_sequence=px.colors.qualitative.Set1 # Use a clear categorical colorscale
                                 )
fig_daily_categorized.update_layout(xaxis={'categoryorder':'category ascending'}) # Ensure chronological order
fig_daily_categorized.show()
```
